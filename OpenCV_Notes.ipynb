**Importing the libraries**

**Shape of the Image**

For example if the shape of the image is (800,800,3) it means that it is 800 pixels wide, and the height of the image would be 800 pixels as well. The 3 corresponds to the number of channels, i.e we know that a picture has three channels, RGB which collectively make up the picture, contributing to the color, resolution, etc. in it's own way

```python
img.shape
```

**Displaying the Image**

We can display the image by using plt.imshow(img). But OpenCV's color channels are in the order of BGR, whereas Matplotlib reads it as RGB. Hence we get a blue-ish image as shown below

```python
plt.imshow(img)
```

To fix the above issue, we use cv2.cvtColor(). This converts the image from it's original colorspace to the colorspace you want it to be

```python
fixed_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #Converts from BGR to RGB
plt.imshow(fixed_img)
```

**Converting an image from color to grayscale**

```python
img_gray = cv2.imread(img_loc,cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

To make it absolutely grayscale we do the following

```python
plt.imshow(img_gray,cmap='gray')
```

**Image Resizing**

```python
img_resize = cv2.resize(fixed_img,(800,400))
plt.imshow(img_resize)
```

Reshaping the image with a specific ratio

```python
x_ratio = 0.5
y_ratio = 0.5
img_spratio = cv2.resize(fixed_img,(0,0),fixed_img,x_ratio,y_ratio)
plt.imshow(img_spratio)
```

The image has been reduced by 50% on the X-axis and Y-axis

**Flipping Images**

We use cv2.flip(img,num)...

if num = 0 ---> Flipped horizontally

if num = 1 ---> Flipped vertically

if num = -1 ---> Simultaneous flip

```python
flipped_img = cv2.flip(fixed_img,-1)
plt.imshow(flipped_img)
```

**Drawing on Images**

We first create an arbitrary blank image

```python
img_blank = np.zeros(shape=(256,256,3),dtype=np.int16)
plt.imshow(img_blank)
```

Drawing a Rectangle

pt1 is the coordinates of the top-left corner, and pt2 is the coordinates of the bottom-right corner

```python
cv2.rectangle(img_blank,pt1=(100,50),pt2=(200,150),color=(255,0,0),thickness=10) #Red Color Line
```

```python
plt.imshow(img_blank)
```

Drawing a circle

```python
cv2.circle(img=img_blank,center=(50,200),radius=25,color=(0,0,255),thickness=10)
```

```python
plt.imshow(img_blank)
```

If we set the thickness=-1, then we get a filled shape

Drawing a Line

```python
cv2.line(img=img_blank,pt1=(0,0),pt2=(255,255),color=(255,211,189),thickness=3)
plt.imshow(img_blank)
```

**Text on Images**

The types of fonts in OpenCV are given in this link: https://www.oreilly.com/library/view/mastering-opencv-4/9781789344912/16b55e96-1027-4765-85d8-ced8fa071473.xhtml

*img* - Image on which the text should be written on

*text* - The text that should be printed

*org* - Bottom Left corner for the text to go into

*fontFace* - Font type

*fontScale* - Like the font size

*color* - Color of the text

*thickness* - Font thickness

*lineType* - Type of the line(s)



```python
font = cv2.FONT_HERSHEY_PLAIN
cv2.putText(img=img_blank,text="GG",org=(50,150),fontFace=font,fontScale=8,color=(255,255,255),thickness=7,lineType=cv2.LINE_AA)
```

```python
plt.imshow(img_blank)
```

**Drawing Custom Polygons**

```python
img_blank = np.zeros(shape=(256,256,3),dtype=np.int16)
plt.imshow(img_blank)
```

Specifying Vertices

```python
vertices = np.array([[25,75],[50,200],[222,222],[230,70]],dtype=np.int32)  
vertices.shape
```

We have to make it into 3 dimensions

```python
pts = vertices.reshape((-1,1,2)) #Adds another set of brackets
```

```python
pts
```

```python
cv2.polylines(img_blank,[pts],isClosed=True,color=(0,255,0),thickness=3)
plt.imshow(img_blank)
```

**Colorspaces**

Along with the RGB(Red Green Blue) colorspace, there are HSL(Hue,Saturation,Lightness), and the HSV(Hue,Saturation,Value) colorspaces.

image.png

```python
#converting from RGB to HSL
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/content/doge.png')
img_modified = cv2.cvtColor(img,cv2.COLOR_RGB2HLS)
plt.imshow(img_modified)
```

```python
#converting from RGB to HSV

import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/content/doge.png')
img_modified = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img_modified)
```
