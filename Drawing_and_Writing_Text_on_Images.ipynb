```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**Drawing on Images**

We first create an arbitrary blank image

```python
img_blank = np.zeros(shape=(256,256,3),dtype=np.int16)
plt.imshow(img_blank)
```

Drawing a Rectangle

pt1 is the coordinates of the top-left corner, and pt2 is the coordinates of the bottom-right corner

```python
cv2.rectangle(img_blank,pt1=(100,50),pt2=(200,150),color=(255,0,0),thickness=10) #Red Color Line, refer to the color scheme for more details
```

```python
plt.imshow(img_blank)
```

Drawing a circle

```python
cv2.circle(img=img_blank,center=(50,200),radius=25,color=(0,0,255),thickness=10)
```

```python
plt.imshow(img_blank)
```

If we set the thickness=-1, then we get a filled shape

Drawing a Line

```python
cv2.line(img=img_blank,pt1=(0,0),pt2=(255,255),color=(255,211,189),thickness=3)
plt.imshow(img_blank)
```

**Text on Images**

The types of fonts in OpenCV are given in this link: https://www.oreilly.com/library/view/mastering-opencv-4/9781789344912/16b55e96-1027-4765-85d8-ced8fa071473.xhtml

The arguments for cv2.putText are as follows:
*img* - Image on which the text should be written on

*text* - The text that should be printed

*org* - Bottom Left corner for the text to go into

*fontFace* - Font type

*fontScale* - Like the font size

*color* - Color of the text

*thickness* - Font thickness

*lineType* - Type of the line(s)


```python
font = cv2.FONT_HERSHEY_PLAIN
cv2.putText(img=img_blank,text="HI",org=(50,150),fontFace=font,fontScale=8,color=(255,255,255),thickness=7,lineType=cv2.LINE_AA)
```

```python
plt.imshow(img_blank)
```

If you change the arguments after it is ran once, then the previous text won't get changed. As you have noticed, I first have written "GG" and then I changed it to "HI". Still, overlapping is visible.

**Drawing Custom Polygons**

```python
img_blank = np.zeros(shape=(256,256,3),dtype=np.int16)
plt.imshow(img_blank)
```

Specifying Vertices

```python
vertices = np.array([[25,75],[50,200],[222,222],[230,70]],dtype=np.int32)  
vertices.shape
```

We have to make it into 3 dimensions

```python
pts = vertices.reshape((-1,1,2)) #Adds another set of brackets
```

```python
pts
```

```python
cv2.polylines(img_blank,[pts],isClosed=True,color=(0,255,0),thickness=3)
plt.imshow(img_blank)
```
