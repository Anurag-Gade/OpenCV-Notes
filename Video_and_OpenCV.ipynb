Firstly, we need to know the procedure of connecting to our camera. It is important to know that opening more than 1 kernel for reading a video stream, will confuse OpenCV in terms of connecting with the cameras, and there would be a conflict of interest.

As the name suggests cv2.VideoCapture does the video capturing from cameras to video files! cv2.VideoWriter is to save to video file. The syntax goes like cv2.VideoWriter(filename, fourcc, fps, frameSize) . The parameter "filename" remains consistent with it's name. "fourcc" is the 4-char code of codec. Codec compresses or decompresses digital video. DIVX is for Windows OS, and similarly DIVX, MJPG, XVID, WMV1 are for the Fedora Distribution of Linux OS. XVID is more used when it comes to Ubuntu, and MacOS. For more details of fourcc see the site https://www.fourcc.org/.
