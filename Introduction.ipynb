**Importing the libraries**

```python
import cv2 #importing the OpenCV library
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
%matplotlib inline
```

**Reading the Image**

cv2.imread(img_path) is used to read the image


```python
img_loc = '/content/doge.png'
img = cv2.imread(img_loc)
```

The type of the image is supposed to be a numpy.ndarray... By any chance if you get a None type even after loading the image, then it might be a wrong path in the argument

```python
type(img)
```

**Shape of the Image**

For example if the shape of the image is (800,800,3) it means that it is 800 pixels wide, and the height of the image would be 800 pixels as well. The 3 corresponds to the number of channels, i.e we know that a picture has three channels, RGB which collectively make up the picture, contributing to the color, resolution, etc. in it's own way.

```python
img.shape
```

**Displaying the Image**

We can display the image by using plt.imshow(img). But OpenCV's color channels are in the order of BGR, whereas Matplotlib reads it as RGB. Hence we get a blue-ish image as shown below

```python
plt.imshow(img)
```

To fix the above issue, we use cv2.cvtColor(). This converts the image from it's original colorspace to the colorspace you want it to be


```python
fixed_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #Converts from BGR to RGB
plt.imshow(fixed_img)
```

**Converting an image from color to grayscale**

```python
img_gray = cv2.imread(img_loc,cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

To make it absolutely grayscale we do the following

```python
plt.imshow(img_gray,cmap='gray')
```

**Image Resizing**

```python
img_resize = cv2.resize(fixed_img,(800,400))
plt.imshow(img_resize)
```

Reshaping the image with a specific ratio

```python
x_ratio = 0.5
y_ratio = 0.5
img_spratio = cv2.resize(fixed_img,(0,0),fixed_img,x_ratio,y_ratio)
plt.imshow(img_spratio)
```

The image has been reduced by 50% on the X-axis and Y-axis

**Flipping Images**

We use cv2.flip(img,num)...

if num = 0 ---> Flipped horizontally

if num = 1 ---> Flipped vertically

if num = -1 ---> Simultaneous flip

```python
flipped_img = cv2.flip(fixed_img,-1)
plt.imshow(flipped_img)
```

**Saving a new copy of the image**

```python
cv2.imwrite('new_version_test.jpg',fixed_img)
```

**Displaying images with OpenCV**

Till here, we have displayed images using plt.imshow(). But it is possible to do the same with the cv2.imshow() method too. ***BUT AFTER USING THIS COMMAND WE MUST USE cv2.waitkey(), WITHOUT THIS THE cv2.imshow() WILL NOT WORK PROPERLY!!!***

Google Colab cannot run the cv2.imshow(img) command unlike the Spyder IDE, so we must use the cv2_imshow(img) command instead. It is imported from google.colab.patches

We get -1 if everything was run correctly


```python
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/doge.png') 
cv2_imshow(img)
cv2.waitKey()
```
