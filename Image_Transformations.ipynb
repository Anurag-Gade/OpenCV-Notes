**Colorspaces**

Along with the RGB(Red Green Blue) colorspace, there are HSL(Hue,Saturation,Lightness), and the HSV(Hue,Saturation,Value) colorspaces.

image.png

```python
#converting from RGB to HSL
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/content/doge.png')
img_modified = cv2.cvtColor(img,cv2.COLOR_RGB2HLS)
plt.imshow(img_modified)
```

```python
#converting from RGB to HSV

import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/content/doge.png')
img_modified = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img_modified)
```

```python
img1 = cv2.imread('/content/walter.jpg')
img2 = cv2.imread('/content/doge.png')
#Reading as RGB
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img1.shape
```

```python
img2.shape
```

```python
#Resizing the images to be of same sizes
img1 = cv2.resize(img1,(430,430))
img2 = cv2.resize(img2,(430,430))
plt.imshow(img2)
```

```python
blended_img = cv2.addWeighted(src1=img1,alpha=0.5,src2=img2,beta=0.5,gamma=0)
```

```python
plt.imshow(blended_img)
```

If we want to increase the dominance of one particular image over the other, then we can tweak the alpha and/or beta values

Blending without Resizing

```python
img1 = cv2.imread('/content/walter.jpg')
img2 = cv2.imread('/content/doge.png')
#Reading as RGB
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
#Mentioning the smaller and larger images
smaller_img = img1
larger_img = img2
#Setting the top left corner's coordinates
x_tl = 0
y_tl = 0
#Ending coordinates
x_ending = x_tl + smaller_img.shape[1]
y_ending = y_tl + smaller_img.shape[0]
```

```python
#Adding the smaller image
larger_img[y_tl:y_ending,x_tl:x_ending] = smaller_img
```

```python
plt.imshow(larger_img)
```
